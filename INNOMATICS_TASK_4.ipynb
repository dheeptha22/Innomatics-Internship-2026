**Problem 1: Mobile Recharge Validation System**

```python
def recharge_validation():
    valid_plans=[199,299,399,599]
    while True:
        amount=int(input("Enter recharge amount: "))
        if amount>=50 and amount in valid_plans:
            print("Recharge Successful")
            break
        else:
            print("Invalid Recharge. Try Again.")
recharge_validation()
```

**Problem 2: Inventory Reorder Alert System**

```python
def inventory_alert(products):
    for name,stock in products.items():
        if stock<15:
            print(name,"Reorder Alert")
        else:
            print(name,"Stock OK")
products={"Rice":20,"Oil":10,"Sugar":5,"Wheat":25}
inventory_alert(products)
```

**Problem 3: Student Result Processing System**

```python
def result_processing(marks):
    total=0
    for m in marks:
        total+=m
    avg=total/len(marks)
    if avg>=50:
        return "Pass"
    else:
        return "Fail"
marks=[60,55,40,70]
print(result_processing(marks))
```

**Problem 4: Cab Fare Estimator with Retry Option**

```python
def cab_fare(km,peak):
    fare=50+(12*km)
    if peak=="yes":
        fare=fare*1.25
    return fare
while True:
    km=float(input("Enter distance in km: "))
    peak=input("Peak hour? (yes/no): ")
    print("Total Fare:",cab_fare(km,peak))
    retry=input("Retry? (yes/no): ")
    if retry!="yes":
        break
```

**Problem 5: Employee Attendance Eligibility Checker**

```python
def attendance_checker(attendance):
    present=0
    for day in attendance:
        if day=="P":
            present+=1
    percentage=(present/len(attendance))*100
    if percentage>=75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance=["P","P","A","P","P"]
print(attendance_checker(attendance))
```

**Problem 6: Password Strength Checker**

```python
def password_checker(password):
    if len(password)<8:
        return "Weak Password"
    digit=False
    special=False
    for ch in password:
        if ch.isdigit():
            digit=True
        if ch in "@#$":
            special=True
    if digit and special:
        return "Strong Password"
    else:
        return "Weak Password"
pwd=input("Enter password: ")
print(password_checker(pwd))
```
