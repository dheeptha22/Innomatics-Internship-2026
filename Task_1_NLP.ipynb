**Task 1:** Conceptual Understanding

**1. Difference between "Love" and "love" in NLP?**

*  "Love" and "love" are treated as different tokens in case-sensitive systems.
*   This increases vocabulary size unnecessarily.
*   That’s why we convert everything to lowercase → normalization.

**2. What happens if stopwords are not removed?**

*   Model gets too much noise
*   Increased computation
*   Reduced accuracy in some tasks (like classification)





**3. Provide two real-world scenarios where removing stopwords can be harmful.**


*   Sentiment Analysis : "I am not happy" → removing "not" changes meaning.


* Question Answering: "What is the capital of India?" → removing "what","is"                        breaks intent






**  4. Explain the difference between stemming and lemmatization.**


1.   Stemming :

               *   Cuts words blindly
               *   Example: "running" → "run"

2.   Lemmatization :


                    *   Uses vocabulary + grammar
                    *   Example: "better" → "good"










**Task 2:** Build Advanced Preprocessing Function


```python
import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return []

    text = text.lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"(.)\1{2,}", r"\1", text)
    text = re.sub(r"[^a-z\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()

    tokens = text.split()
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    return tokens
text = "I absolutely looooved this product 😍😍"

result = preprocess_text(text)

print(result)
```

**Task 3:** Stress Testing

```python
# Sample Inputs
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Run preprocessing on each sentence
for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    cleaned_sentence = " ".join(tokens)

    print("Original Text: ", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print("-" * 60)
```

**Task 4:** Token Analytics


```python
print("\n===== TOKEN ANALYTICS =====\n")

most_noise_sentence = ""
max_removed = -1

most_meaningful_sentence = ""
max_tokens = -1

for sentence in test_sentences:
    original_words = sentence.split()

    tokens = preprocess_text(sentence)
    cleaned_sentence = " ".join(tokens)

    # Metrics
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_token_length = (
        sum(len(word) for word in tokens) / total_tokens
        if total_tokens > 0 else 0
    )

    # Noise estimation (how many words removed)
    removed_count = len(original_words) - total_tokens

    if removed_count > max_removed:
        max_removed = removed_count
        most_noise_sentence = sentence

    if total_tokens > max_tokens:
        max_tokens = total_tokens
        most_meaningful_sentence = sentence

    # Output per sentence
    print("Original Text:", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned_sentence)

    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_token_length, 2))
    print("-" * 60)
```

**Task 5:** Frequency Analysis

```python
from collections import Counter

# Collect all tokens from all sentences
all_tokens = []

for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    all_tokens.extend(tokens)

# Create frequency counter
counter = Counter(all_tokens)

# Top 10 most frequent words
top_10 = counter.most_common(10)

# Top 5 least frequent words
least_5 = sorted(counter.items(), key=lambda x: x[1])[:5]

print("===== FREQUENCY ANALYSIS =====\n")

print("Top 10 Most Frequent Words:")
for word, freq in top_10:
    print(f"{word} : {freq}")

print("\nTop 5 Least Frequent Words:")
for word, freq in least_5:
    print(f"{word} : {freq}")
```

**Task 6:** Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)   # reuse your function
        clean_sentence = " ".join(tokens)

        all_tokens.extend(tokens)
        clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
result = full_pipeline(test_sentences)

print("Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
for s in result["clean_sentences"]:
    print(s)
```

**Task 7:** Error Handling

```python
import re

def preprocess_text(text):
    # Handle invalid input
    if not text or not isinstance(text, str):
        return []

    # Convert to lowercase
    text = text.lower()

    # Remove URLs and emails
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Normalize repeated characters
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove special characters (including emojis)
    text = re.sub(r"[^a-z\s]", "", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # If text becomes empty after cleaning
    if not text:
        return []

    # Tokenize
    tokens = text.split()

    # Remove short tokens (keep "no", "not")
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    return tokens
edge_cases = [
    "",                 # Empty string
    "😂😂😂😂😂",         # Only emojis
    "1234567890",       # Only numbers
    "!!!@@@",           # Only symbols
    "   "               # Only spaces
]

print("===== EDGE CASE TESTING =====\n")

for case in edge_cases:
    tokens = preprocess_text(case)
    cleaned_sentence = " ".join(tokens)

    print("Input:", repr(case))
    print("Tokens:", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print("-" * 50)
```
