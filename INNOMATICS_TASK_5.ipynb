**Problem 1: Social Media – Post Engagement Analyzer**

```python
def analyze_engagement(likes_list):
    total_likes=0
    for likes in likes_list:
        total_likes+=likes
    if total_likes>=1000:
        status="Viral Post"
    else:
        status="Normal Engagement"
    print("Total Likes:",total_likes)
    print("Post Status:",status)

likes_data=[200,300,150,400]
analyze_engagement(likes_data)
```

**Problem 3: Agriculture – Rainfall Adequacy Checker**

```python
def rainfall_checker(rainfall,required):
    total=0
    for r in rainfall:
        total+=r
    avg=total/len(rainfall)
    if avg>=required:
        status="Adequate Rainfall"
    else:
        status="Inadequate Rainfall"
    print("Average Rainfall:",int(avg))
    print("Rainfall Status:",status)

rainfall_checker([70,75,65,78],70)
```

**Problem 4: Social Media – Duplicate Account Detection**

```python
def duplicate_accounts(usernames):
    if len(usernames)!=len(set(usernames)):
        result="Yes"
    else:
        result="No"
    print("Duplicate Accounts Found:",result)

duplicate_accounts(["user1","user2","user1","user3"])
```

**Problem 6: Agriculture – Premium Crop Price Filter**


```python
def premium_crops(prices):
    premium=[]
    for price in prices:
        if price>2000:
            premium.append(price)
    print("Premium Crops:",premium)

premium_crops([1500,2500,1800,3200])
```

**Problem 7: System Monitoring – Application Health Checker**

```python
def app_health(errors):
    if errors==0:
        status="Healthy"
    elif errors<=5:
        status="Minor Issues"
    else:
        status="Critical Issues"
    print("Error Count:",errors)
    print("System Status:",status)

app_health(7)
```

**Problem 9: E-Learning – Student Attendance Eligibility System**

```python
def attendance_eligibility(attendance):
    total=0
    for a in attendance:
        total+=a
    percentage=(total/len(attendance))
    if percentage>=75:
        status="Eligible"
    else:
        status="Not Eligible"
    print("Attendance Percentage:",percentage)
    print("Exam Eligibility:",status)

attendance_eligibility([80,85,75,80])
```

**Problem 10: Smart Electricity Bill Analyzer**

```python
def electricity_bill(units):
    if units<=100:
        bill=units*3
    elif units<=200:
        bill=100*3+(units-100)*5
    else:
        bill=100*3+100*5+(units-200)*7
    if bill<500:
        usage="Low Usage"
    elif bill<=1500:
        usage="Moderate Usage"
    else:
        usage="High Usage"
    return bill,usage

bill,usage=electricity_bill(250)
print("Total Bill:",bill)
print("Usage Status:",usage)
```
